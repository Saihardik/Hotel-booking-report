```python

df = df.drop(['company','agent'],axis = 1,errors='ignore')
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df['adr'].plot(kind='box')
```

```python
# there was an outlier at 5400

df = df[df['adr']<5000]
```

```python
df['adr'].plot(kind='box')
```

```python
# finding the cancelling percentage


cancel_percent = df['is_canceled'].value_counts(normalize = True)
print(cancel_percent)
counts = df['is_canceled'].value_counts()


# creating the visuals

plt.figure(figsize=(5,4))
plt.title('Reservation Status count')
bars = plt.bar(['Not cancelled',"Cancelled"],df['is_canceled'].value_counts(),edgecolor='blue',width=0.3)

# for bar in bars:
#     height = bar.get_height()
#     plt.text(bar.get_x() + bar.get_width()/2, height + 0.01 * max(counts),
#              f'{int(height)}', ha='center', va='bottom', fontsize=10)

plt.ylabel("Count")
plt.show()
```

```python
# visualizing counts based in the hotels

plt.figure(figsize=(6,4))

bars = sns.countplot(x='hotel',hue='is_canceled',data = df,palette=['green','red'],width=0.5)

legend_lables = bars.get_legend_handles_labels()

bars.legend(bbox_to_anchor=(1,1))

plt.title('Reservation status on different hotels',size=20)

plt.xlabel('hotels')

plt.ylabel('No. Of reservations')

plt.legend(['not canceled','canceled'])

plt.show()

```

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(14,6))
plt.title(' Average daily rate in city and resort hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='resort hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='city hotel')
plt.legend(fontsize=20)
plt.show()

```

```python
 Visulaizing the cancellations and non concellations on every month
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
plt.title('Reservation status per month')
ax1 = sns.countplot(x='month',hue='is_canceled',data=df,palette=['green','red'])
plt.xlabel('month')
plt.ylabel('no of reservations')
plt.legend(['not canceled','canceled'])

plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=20)
sns.barplot(x='month',y='adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index(),color='blue')
plt.show()
```

```python
cancelled_data = df[df['is_canceled']==1]
top_10_country = cancelled_data['country'].value_counts()[:10]

plt.figure(figsize=(4,4))
plt.title(' Top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct = '%.2f',labels=top_10_country.index)
plt.show()
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_data = df[df['is_canceled']==0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)


plt.figure(figsize=(16,6))
plt.title('Average Daily rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend()
plt.show()
```
